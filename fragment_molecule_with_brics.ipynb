```python
from rdkit import Chem
from rdkit.Chem.BRICS import BRICSDecompose
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
```

```python
mol = Chem.MolFromSmiles('O=c1nc([nH]c2NCC(Nc12)CN(c1ccc(cc1)C(=O)NC(CCC(=O)O)C(=O)O)C=O)N')
```

```python
mol
```

```python
res = list(BRICSDecompose(mol,returnMols=True))
smis = [Chem.MolToSmiles(x,True) for x in res]
```

```python
smis
```

```python
img = Draw.MolsToGridImage(res)
img
```

```python
drugbank_input = Chem.SDMolSupplier('drugbank_approved_structures.sdf')
drugbank = [m for m in drugbank_input if m]
drugbank_scaffolds_list = [list(BRICSDecompose(mol,returnMols=True)) for mol in drugbank]
scaffold_smiles = [Chem.MolToSmiles(scaffold, True) for drugbank_scaffolds in drugbank_scaffolds_list for scaffold in drugbank_scaffolds if scaffold != None]
```

```python
import collections
counter=collections.Counter(scaffold_smiles)
print(counter.most_common(5))
```
